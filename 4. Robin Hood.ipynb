```python
#Varibale assignment
points = [(4, 5), (0, 2), (4, 7), (1, -3), (3, -2), (4, 5), (3, 2), (5, 7), (-5, 7), (2, 2), (-4, 5), (0, -2), (-4, 7), (-1, 3), (-3, 2), (-4, -5), (-3, 2), (5, 7), (5, 7), (2, 2), (9, 9), (-8, -9)]

a = 0
b = 0
c = 0
d = 0
e = 0
f = 0
g = 0

same_point = []
forest = []
distances = []
shortest_distances = []

Q1 = []
Q2 = []
Q3 = []
Q4 = []

#------------------------------------------------------------------

# 1. Robin Hood is famous for hitting an arrow with another arrow. Did you get it?
# 2. Calculate how many arrows have fallen in each quadrant.
# 3. Find the point closest to the center. Calculate its distance to the center.
# 4. If the target has a radius of 9, calculate the number of arrows that must be picked up in the forest.

#------------------------------------------------------------------

#Finding if I hit the same point twice
for i in points:
  temp_list = []
  for i in points:
    if points[a] == points[b]:
      temp_list.append(points[b])
      b += 1
    elif points[a] != points[b]:
      b += 1
    elif ((points.index(points[b - 1])) + 1) == len(points):
      break
  if ((points.index(points[b - 1])) + 1) == len(points):
    a += 1
    b -= len(points)
  if len(temp_list) > 1:
    if (temp_list[0], len(temp_list)) not in same_point:
      same_point.append((temp_list[0], len(temp_list)))

#Printing same points results
print("\nPart 1: \n")
for i in same_point:
  print("Point", same_point[c][0], "was hit", same_point[c][1], "times")
  c += 1

#Calculating how many arrows have fallen in each quadrant
for i in points:
  if points[d][0] > 0 and points[d][1] > 0:
    Q1.append(points[d])
  elif points[d][0] < 0 and points[d][1] > 0:
    Q2.append(points[d])
  elif points[d][0] < 0 and points[d][1] < 0:
    Q3.append(points[d])
  elif points[d][0] > 0 and points[d][1] < 0:
    Q4.append(points[d])
  d += 1

print("\nPart 2: \n")
print("There are", len(Q1) ,"arrows in Q1")
print("There are", len(Q2) ,"arrows in Q2")
print("There are", len(Q3) ,"arrows in Q3")
print("There are", len(Q4) ,"arrows in Q4")

#Closest point to the center 
print("\nPart 3: \n")

for i in points:
  result = ((((points[f][0] - 0)**2)+((points[f][1] - 0)**2))**(1/2))
  distances.append(result)
  f += 1

for i in distances:
  if ((((points[g][0] - 0)**2)+((points[g][1] - 0)**2))**(1/2)) == min(distances):
    shortest_distances.append(points[g])
    g += 1
  elif ((((points[g][0] - 0)**2)+((points[g][1] - 0)**2))**(1/2)) != min(distances):
    g += 1

if len(shortest_distances) == 1:
  print("The arrow at point", str(shortest_distances[0]), "has the shortest distance to the center of the target, it is", min(distances), "unit away from the center") 
elif len(shortest_distances) > 1:
  print("The arrows at points", str(shortest_distances), "have the shortest distance to the center of the target, they are", min(distances), "units away from the center")

#Arrows picked up from the forest floor calculation
print("\nPart 4: \n")
for i in points:
  if abs(points[e][0]) > 9 or abs(points[e][1]) > 9:
    forest.append(points[e])
  e += 1

if len(forest) == 0:
  print("No arrows have to be picked up from the forest floor")
elif len(forest) == 1:
  print(len(forest), "arrow has to be picked up from the forest floor")
elif len(forest) > 1:
  print(len(forest), "arrows have to be picked up from the forest floor")
```
